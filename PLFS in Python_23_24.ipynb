```python
# WPR, LFPR and UR by rural-urban and sex
```

```python
emp_u_total = plfs_23_24_v1.groupby(['Sector', 'Sex']).agg(
    total_pop=('Final Weight', "sum"),
    employed_pop=('Final Weight', lambda x: (x * plfs_23_24_v1.loc[x.index, 'employed']).sum()),
    unemployed_pop=('Final Weight', lambda x: (x * plfs_23_24_v1.loc[x.index, 'unemployed']).sum())
)

emp_u_total
```

```python
emp_u_total["WPR"] = round((emp_u_total['employed_pop'] / emp_u_total['total_pop']) * 100,1)

emp_u_total # exact match with the reports 
```

```python
emp_u_total["LFPR"] = round(((emp_u_total['employed_pop'] + emp_u_total['unemployed_pop']) / emp_u_total['total_pop']) * 100,1)

emp_u_total 
```

```python
emp_u_total["UR"] = round((emp_u_total['unemployed_pop'] / (emp_u_total['employed_pop'] + emp_u_total['unemployed_pop'])) * 100,1)

emp_u_total # exact match with the reports
```

```python
overall = plfs_23_24_v1.agg(
    total_pop=("Final Weight", "sum"),
    employed_pop=("Final Weight", lambda x: (x * plfs_23_24_v1.loc[x.index, "employed"]).sum()),
    unemployed_pop=("Final Weight", lambda x: (x * plfs_23_24_v1.loc[x.index, "unemployed"]).sum()),
)

overall = overall.T

overall
```

```python
overall["WPR"] = round((overall['employed_pop'] / overall['total_pop']) * 100,1)

overall
```

```python
overall["LFPR"] = round(((overall['employed_pop'] + overall['unemployed_pop']) / overall['total_pop']) * 100,1)

overall
```

```python
overall["UR"] = round((overall['unemployed_pop'] / (overall['employed_pop'] + overall['unemployed_pop'])) * 100,1)

overall # exact match with the reports
```
